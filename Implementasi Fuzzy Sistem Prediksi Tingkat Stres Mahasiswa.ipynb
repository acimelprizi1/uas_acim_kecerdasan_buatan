Implementasi Fuzzy Sistem Prediksi Tingkat Stres Mahasiswa

```python
import numpy as np
import matplotlib.pyplot as plt
```

Fungsi keanggotaan untuk jumlah tugas

```python
def sedikit_tugas(x):
    return np.where(x <= 5, 1, np.where(x <= 10, (10 - x) / 5, 0))

def banyak_tugas(x):
    return np.where(x >= 10, 1, np.where(x >= 5, (x - 5) / 5, 0))
```

Fungsi keanggotaan untuk lama waktu belajar

```python
def cukup_belajar(x):
    return np.where(x <= 3, 1, np.where(x <= 6, (6 - x) / 3, 0))

def terlalu_lama_belajar(x):
    return np.where(x >= 6, 1, np.where(x >= 3, (x - 3) / 3, 0))
```

Fungsi keanggotaan untuk lama waktu tidur

```python
def cukup_tidur(x):
    return np.where(x >= 6, 1, np.where(x >= 4, (x - 4) / 2, 0))

def kurang_tidur(x):
    return np.where(x < 6, 1, np.where(x < 8, (8 - x) / 2, 0))
```

Visualisasi fungsi keanggotaan

```python
x_tugas = np.linspace(0, 20, 100)
x_belajar = np.linspace(0, 10, 100)
x_tidur = np.linspace(0, 10, 100)

plt.figure(figsize=(12, 8))
```

Jumlah Tugas

```python
plt.subplot(3, 1, 1)
plt.plot(x_tugas, sedikit_tugas(x_tugas), label="Sedikit Tugas")
plt.plot(x_tugas, banyak_tugas(x_tugas), label="Banyak Tugas")
plt.title("Fungsi Keanggotaan Jumlah Tugas")
plt.xlabel("Jumlah Tugas")
plt.ylabel("Derajat Keanggotaan")
plt.legend()
```

Lama Waktu Belajar

```python
plt.subplot(3, 1, 2)
plt.plot(x_belajar, cukup_belajar(x_belajar), label="Cukup Belajar")
plt.plot(x_belajar, terlalu_lama_belajar(x_belajar), label="Terlalu Lama Belajar")
plt.title("Fungsi Keanggotaan Lama Waktu Belajar")
plt.xlabel("Lama Waktu Belajar (Jam)")
plt.ylabel("Derajat Keanggotaan")
plt.legend()
```

Lama Waktu Tidur

```python
plt.subplot(3, 1, 3)
plt.plot(x_tidur, cukup_tidur(x_tidur), label="Cukup Tidur")
plt.plot(x_tidur, kurang_tidur(x_tidur), label="Kurang Tidur")
plt.title("Fungsi Keanggotaan Lama Waktu Tidur")
plt.xlabel("Lama Waktu Tidur (Jam)")
plt.ylabel("Derajat Keanggotaan")
plt.legend()
```

```python
plt.tight_layout()
plt.show()
```

Aturan Fuzzy (Fuzzy Rules)

```python
def prediksi_stres(jumlah_tugas, waktu_belajar, waktu_tidur):
    tugas = max(sedikit_tugas(jumlah_tugas), banyak_tugas(jumlah_tugas))
    belajar = max(cukup_belajar(waktu_belajar), terlalu_lama_belajar(waktu_belajar))
    tidur = max(cukup_tidur(waktu_tidur), kurang_tidur(waktu_tidur))

    # Derajat stres berdasarkan aturan fuzzy
    stres = min(tugas, belajar, tidur) * 100  # Tingkat stres dalam persen
    return stres
```

Contoh penggunaan

```python
jumlah_tugas = 12  # Banyak tugas
waktu_belajar = 7  # Terlalu lama belajar
waktu_tidur = 4   # Kurang tidur
```

Prediksi tingkat stres

```python
stres = prediksi_stres(jumlah_tugas, waktu_belajar, waktu_tidur)
print(f"Tingkat stres mahasiswa: {stres}%")
```
